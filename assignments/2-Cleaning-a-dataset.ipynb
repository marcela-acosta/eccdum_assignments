## Unifying Victoria's Secret

```python
def unify_victoria_secret(df):
    """
    We want that all brands that are related to Victoria's Secret
    have `victoria's secret` as their brand instead of what they
    currently have.
    """
    df = df.copy()
    new_string = "victoria's secret"

    # Write your code here

    # Verificación de la existencia de la columna y unificación
    if 'brand_name' in df.columns:
        df.loc[df["brand_name"].str.contains("victoria's secret", case=False, na=False), "brand_name"] = new_string
    else:
        print("Advertencia: columna 'brand_name' no encontrada en el DataFrame.")

    return df
```

```python
df_unified = unify_victoria_secret(df)
```

```python
answer_victoria_secret = df_unified[df_unified["brand_name"] == "victoria's secret"].shape[0]
print(answer_victoria_secret)
```

## Cleaning up the price

In this section we are going to transform the `price` column into a float column in USD dolars.

For this, be careful of the different formats in the data.

For simplicity, you might assume that all the prices are in USD dolars, regarding of the symbol of the currency used.

```python
def clean_price(df):
    """
    In this function we will transform the
    `price` column into a column of floats.
    In case a product has more than one price,
    return the lowest one.
    """

    # Write your code here

    df = df.copy()

    # Función para procesar cada precio y convertirlo a float
    def parse_price(price_str):
        # Verificar si el valor es NaN
        if pd.isna(price_str):
            return np.nan

        # Convertir el valor a string, eliminando comas
        price_str = str(price_str).replace(",", "")
        price_list = []

        # Dividir si hay múltiples precios separados por "|"
        for segment in price_str.split("|"):
            try:
                # Limpiar cada precio eliminando símbolos y tomando solo el valor numérico inicial
                price = float(segment.split()[0].replace("$", ""))
                price_list.append(price)
            except (ValueError, IndexError):
                # Ignorar segmentos que no se puedan convertir a número
                continue

        # Retornar el precio más bajo si hay valores válidos, o NaN en caso contrario
        return min(price_list) if price_list else np.nan

    # Aplicar la función de limpieza a la columna `price`
    df["price"] = df["price"].apply(parse_price)

    return df
```

```python
df_clean = clean_price(df_unified)
```

```python
answer_unified_price_sum = df_clean["price"].sum()
print(answer_unified_price_sum)
```

```python
def low_high_product_mean(df):
    """
    Finally, we will calculate `product_category` with the lowest and highest mean price
    for the brand Victoria's Secret.
    """

    # Write your code here

    # Filtrar solo los productos de la marca 'Victoria's Secret'
    vs_df = df[df['brand_name'].str.lower() == "victoria's secret"]

    # Calcular el precio promedio por categoría de producto
    category_mean = vs_df.groupby('product_category')['price'].mean()

    # Encontrar los precios promedio más bajo y más alto
    lowest_mean_price = category_mean.min()
    highest_mean_price = category_mean.max()

    # Retornar solo los valores de los precios
    return lowest_mean_price, highest_mean_price
```
